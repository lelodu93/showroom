```python
from google_play_scraper import reviews_all, Sort
import pandas as pd
```

```python
#ID de l'application Showroom Privé
app_id = "com.holosfind.showroom"

#Récupérer tous les avis
reviews = reviews_all(
    app_id,
    lang='fr',  # Français
    country='fr',  # France
    sort=Sort.MOST_RELEVANT  # Trier par pertinence (ou Sort.NEWEST pour les plus récents)
)
```

```python
reviews #lire les commentaires
```

```python
#Convertir en DataFrame
df = pd.DataFrame(reviews)
```

```python
df.to_excel('avis_showroom.xlsx', index=False) #Exporter les données au format excel
```
