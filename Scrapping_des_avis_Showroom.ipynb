```python
from google_play_scraper import reviews_all, Sort
import pandas as pd
```

```python
#ID de l'application Showroom Priv√©
app_id = "com.holosfind.showroom"

#R√©cup√©rer tous les avis
reviews = reviews_all(
    app_id,
    lang='fr',  # Fran√ßais
    country='fr',  # France
    sort=Sort.MOST_RELEVANT  # Trier par pertinence (ou Sort.NEWEST pour les plus r√©cents)
)
```

```python
reviews #lire les commentaires
```

```python
#Convertir en DataFrame
df = pd.DataFrame(reviews)
```

```python
df.to_excel('avis_showroom.xlsx', index=False) #Exporter les donn√©es au format excel
```
